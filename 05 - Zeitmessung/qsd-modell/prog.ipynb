```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'grid'])
```

```python
timing_ann = np.loadtxt("timing_ann.dat", delimiter=";")
timing_numeric = np.loadtxt("timing_numeric.dat", delimiter=";")
```

```python
plt.figure(figsize=(9,5))
plt.plot(timing_ann[:,0], timing_ann[:,3], color="blue", label="neuronales Netz")
plt.plot(timing_numeric[:,0], timing_numeric[:,1], color="red" ,label="numerische Berechnung")
plt.legend(fontsize=11)
plt.xlabel("Anzahl der Sites", fontsize=12)
plt.ylabel("Zeit (s)", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig("T_sqd.png", dpi=200, format="png")
plt.show()
```
