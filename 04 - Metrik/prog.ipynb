```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'grid'])
```

```python
s_abs_3 = np.loadtxt("s_abs.dat", delimiter=";")
s_abs_10 = np.loadtxt("s_abs_10.dat", delimiter=";")
s_abs_qsd3 = np.loadtxt("s_abs_qsd_3.dat", delimiter=";")
num_iter = len(s_abs_3)
```

```python
plt.figure(figsize=(9,5))
plt.plot(np.arange(0, num_iter), s_abs_3, color="blue", label="3-Site-Modell", linewidth=1.5)
plt.plot(np.arange(0, num_iter), s_abs_10, color="red", label="10-Site-Modell", linewidth=1.5)
plt.plot(np.arange(0, num_iter), s_abs_qsd3, color="green", label="3-Site-QSD-Modell", linewidth=1.5)
plt.xlabel("Anzahl der Iterationen", fontsize=12)
plt.ylabel("$|S|$", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=11)
plt.savefig("S_abs.png", dpi=300, format="png")
plt.show()
```
